```python
import random
import csv
from matplotlib import pyplot as plt

file=open('userdata.csv','w')
csvwriter = csv.writer(file) 
fields=["1","2"]
csvwriter.writerow(fields)

li1=[]
li2=[]

for i in range(10):
  value=random.randint(1,10)
  print(value)
  li1.append(value)

for i in range(10):
  value=random.randint(100,500)
  li2.append(value)

for i in range(10):
  plt.plot(li1[i-1],li2[i-1])
  row=[li1[i-1],li2[i-1]]
  csvwriter.writerow(row) 

plt.plot([li1[0],li2[0]],[li1[1],li2[1]],[li1[2],li2[2]],[li1[3],li2[3]],[li1[4],li2[4]],[li1[5],li2[5]],[li1[6],li2[6]],[li1[7],li2[7]],[li1[8],li2[8]],[li1[9],li2[9]])

file.close()
```
